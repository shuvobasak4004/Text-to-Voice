```python
from gtts import gTTS
from IPython.display import Audio
from google.colab import output

def text_to_speech():
    text = input("Enter the text to convert to speech (Bengali): ")
    language = 'bn'  # Set the language code to 'bn' for Bengali
    
    tts = gTTS(text=text, lang=language)
    tts.save('output.mp3')
    
    audio = Audio('output.mp3')
    output.clear()
    display(audio)

# Call the function to convert text to speech
text_to_speech()
```
